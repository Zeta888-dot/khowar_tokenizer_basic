```python
from tokenizers import ByteLevelBPETokenizer
from transformers import PreTrainedTokenizerFast
from huggingface_hub import login
import os
```

```python
login(token="YOUR_HUGGING_FACE_TOKEN")
```

```python
raw_corpus = "khowar_dataset.jsonl"
clean_corpus = "khowar_utf8.jsonl"

with open(raw_corpus, "rb") as f:
    text = f.read().decode("utf-8", errors="ignore")

with open(clean_corpus, "w", encoding="utf-8") as f:
    f.write(text)

print("UTF-8 corpus ready")
```

```python
tokenizer = ByteLevelBPETokenizer()

tokenizer.train(
    files=[clean_corpus],
    vocab_size=1500,
    min_frequency=2,
    special_tokens=["<s>", "<pad>", "</s>", "<unk>", "<mask>"]
)
print("Tokenizer trained")
```

```python
tokenizer_dir = "khowar_tokenizer"
os.makedirs(tokenizer_dir, exist_ok=True)

tokenizer.save(f"{tokenizer_dir}/tokenizer.json")

print("tokenizer.json saved")
```

```python
hf_tokenizer = PreTrainedTokenizerFast(
    tokenizer_file=f"{tokenizer_dir}/tokenizer.json",
    bos_token="<s>",
    eos_token="</s>",
    unk_token="<unk>",
    pad_token="<pad>",
    mask_token="<mask>",
)
hf_tokenizer.model_max_length = 1024
hf_tokenizer.save_pretrained(tokenizer_dir)

print("PreTrainedTokenizerFast ready")
```

```python
text = "aba hash no boy la"
enc = hf_tokenizer(text)

print(enc)
print(hf_tokenizer.decode(enc["input_ids"]))
```
